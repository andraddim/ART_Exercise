```python
import random
import matplotlib.pyplot as plt
```

```python
def random_five():
    num=random.randint(1,5)
    return num
```

```python

num_occurrences=[0,0,0,0,0,0,0]
for i in range(1000):
    sum=0

    for i in range(7):
        sum+=random_five()

    result=1+sum%7
    num_occurrences[result-1]+=1

probabilities=[occurrences/1000 for occurrences in num_occurrences]

print("Number of occurrences:")
for i in range(7):
    print("Number "+str(i+1)+" appears "+str(num_occurrences[i])+ " times")


plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(range(1, 8), num_occurrences)
plt.title('Occurrences of each number')
plt.xlabel('Number')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)
plt.bar(range(1, 8), probabilities)
plt.title('Probability of each number')
plt.xlabel('Number')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```
